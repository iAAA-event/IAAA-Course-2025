```python
# @title Section 1: Coin Flip Test
from IPython.display import HTML

HTML('''
<div style="padding: 12px; background: linear-gradient(to right, #c471ed, #f7797d); color: white; font-size: 22px; border-radius: 8px; margin-top: 20px;">
  üìò Section 1: Coin Flip Test
</div>
''')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
num_trials = 1000  # Number of coin tosses
coin_sides = ["Heads", "Tails"]

# Simulate coin tosses
tosses = np.random.choice(coin_sides, num_trials)

# Count occurrences of each outcome
heads_count = np.sum(tosses == "Heads")
tails_count = np.sum(tosses == "Tails")

# Compute empirical probabilities
empirical_probabilities = [heads_count / num_trials, tails_count / num_trials]
print(empirical_probabilities)

# Create bar chart
fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(coin_sides, empirical_probabilities, color=["blue", "green"], alpha=0.7)

# Labeling the plot
ax.set_ylabel("Empirical Probability")
ax.set_title(f"Probability Space for {num_trials} Coin Tosses")
ax.set_ylim(0, 1)
ax.grid(axis="y", linestyle="--", alpha=0.5)

# Show the plot
plt.show()
```

```python
# @title Section 2: Dice Roll Test
from IPython.display import HTML

HTML('''
<div style="padding: 12px; background: linear-gradient(to right, #c471ed, #f7797d); color: white; font-size: 22px; border-radius: 8px; margin-top: 20px;">
  üìò Section 2: Dice Roll Test
</div>
''')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
num_rolls = 10000  # Number of die rolls
die_faces = np.arange(1, 7)  # Possible outcomes: 1 to 6

# Simulate die rolls
rolls = np.random.choice(die_faces, num_rolls)

# Count occurrences of each face
counts = np.bincount(rolls, minlength=7)[1:]  # Ignore index 0

# Compute empirical probabilities
empirical_probabilities = counts / num_rolls

# Create bar chart
fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(die_faces, empirical_probabilities, color="blue", alpha=0.7)

# Labeling the plot
ax.set_xlabel("Die Face")
ax.set_ylabel("Empirical Probability")
ax.set_title(f"Probability Space for {num_rolls} Rolls")
ax.set_ylim(0, 1)
ax.grid(axis="y", linestyle="--", alpha=0.5)

# Show the plot
plt.show()
```

```python
# @title Section 4: Two coins flip
from IPython.display import HTML

HTML('''
<div style="padding: 12px; background: linear-gradient(to right, #c471ed, #f7797d); color: white; font-size: 22px; border-radius: 8px; margin-top: 20px;">
  üìò Section 4: Two coins flip
</div>
''')
```

```python
import random
import matplotlib.pyplot as plt

def toss_two_coins():
    # Toss two coins: each can be 'Heads' or 'Tails'
    coin1 = random.choice(['H', 'T'])
    coin2 = random.choice(['H', 'T'])
    return coin1, coin2

# Simulate the tossing of two coins 1000 times
n_tosses = 100
outcomes = {'HH': 0, 'HT': 0, 'TH': 0, 'TT': 0}

# Simulating the experiment
for _ in range(n_tosses):
    result = toss_two_coins()
    outcome = ''.join(result)  # Combine the result into a string (e.g., "HT")
    outcomes[outcome] += 1

# Plotting: Bar Chart for the frequencies of outcomes
labels = list(outcomes.keys())
values = list(outcomes.values())
values = [v/n_tosses for v in values]

# Bar chart to show frequency of each outcome
plt.figure(figsize=(8, 5))
plt.bar(labels, values, color=['red', 'green', 'blue', 'orange'])
plt.title(f"Outcome Frequencies for Tossing Two Coins ({n_tosses} Simulations)")
plt.xlabel("Outcomes")
plt.ylabel("Frequency")
plt.show()

# Optional: Pie chart for visualizing the proportions of outcomes
plt.figure(figsize=(7, 7))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=['red', 'green', 'blue', 'orange'])
plt.title(f"Outcome Proportions for Tossing Two Coins ({n_tosses} Simulations)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1D Normal Distribution Visualization
def plot_1d_normal(mu=0, sigma=1, num_points=1000):
    # Generate values for the x-axis
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, num_points)

    # Get the corresponding y values of the PDF of the normal distribution
    y = norm.pdf(x, mu, sigma)

    # Plot the 1D normal distribution
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=f'Normal Distribution\n$\mu={mu}$, $\sigma={sigma}$', color='blue')
    plt.fill_between(x, y, color='blue', alpha=0.2)
    plt.title("1D Normal Distribution")
    plt.xlabel("X")
    plt.ylabel("Probability Density")
    plt.grid(True)
    plt.legend()
    plt.show()


# Visualize 1D Normal Distribution with mean 0 and std dev 1
plot_1d_normal(mu=0, sigma=1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import multivariate_normal

# Parameters for the bivariate normal distribution
mu = [0, 0]  # Mean vector (mu_x, mu_y)
sigma = [[1, 0.5],  # Covariance matrix (sigma_xx, sigma_xy)
         [0.5, 1]]  # (sigma_yx, sigma_yy)

# Create a grid of (x, y) values
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

# Stack the X and Y values into a (n, 2) array
pos = np.dstack((X, Y))

# Create the bivariate normal distribution
rv = multivariate_normal(mu, sigma)

# Calculate the PDF values for each point in the (x, y) grid
Z = rv.pdf(np.dstack((X, Y)))

# Plotting the 3D surface of the bivariate normal distribution
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Density')
ax.set_title('3D Bivariate Normal Distribution')

# Add color bar
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 2D Normal Distribution Visualization
def plot_2d_normal(mu_x=0, mu_y=0, sigma_x=1, sigma_y=1, rho=0, num_points=100):
    # Create grid of (x, y) values
    x = np.linspace(mu_x - 4*sigma_x, mu_x + 4*sigma_x, num_points)
    y = np.linspace(mu_y - 4*sigma_y, mu_y + 4*sigma_y, num_points)
    X, Y = np.meshgrid(x, y)

    # Compute the bivariate normal distribution (PDF)
    Z = np.exp(-0.5 * ((X - mu_x)**2 / sigma_x**2 + (Y - mu_y)**2 / sigma_y**2)) / (2 * np.pi * sigma_x * sigma_y)

    # Plot the 2D normal distribution using contour plot
    plt.figure(figsize=(8, 6))
    cp = plt.contour(X, Y, Z, levels=10, cmap='Blues')
    # plt.title(f"2D Normal Distribution\n$\mu_x={mu_x}$, $\mu_y={mu_y}$, $\sigma_x={sigma_x}$, $\sigma_y={sigma_y}$, $\rho={rho}$")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.colorbar(cp)
    plt.grid(True)
    plt.show()

# Visualize 2D Normal Distribution with mean (0,0) and std dev (1,1)
plot_2d_normal(mu_x=0, mu_y=0, sigma_x=1, sigma_y=1, rho=0)
```

```python
# @title Section 6: Bernoulli Distribution
from IPython.display import HTML

HTML('''
<div style="padding: 12px; background: linear-gradient(to right, #c471ed, #f7797d); color: white; font-size: 22px; border-radius: 8px; margin-top: 20px;">
  üìò Section 6: Bernoulli Distribution
</div>
''')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to simulate Bernoulli trials
def simulate_bernoulli(p, n):
    # p is the probability of success, n is the number of trials
    return np.random.binomial(1, p, n)

# Parameters
p = 0.7  # Probability of success (can be adjusted)
n = 1000  # Number of trials

# Simulate the Bernoulli distribution
bernoulli_trials = simulate_bernoulli(p, n)


# Plotting the probability distribution (Bar plot)
plt.figure(figsize=(8, 6))
plt.bar([0, 1], [np.mean(bernoulli_trials == 0), np.mean(bernoulli_trials == 1)], color=['red', 'green'])
plt.title(f'Bernoulli Probability Distribution (p={p})')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.xticks([0, 1], ['Failure (0)', 'Success (1)'])
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the exponential distribution
lambda_ = .5  # Rate parameter (1 / mean)
size = 1000  # Number of simulations

# Simulate exponential distribution
exponential_data = np.random.exponential(1/lambda_, size)

# Plot the histogram
plt.figure(figsize=(8, 6))
plt.hist(exponential_data, bins=30, density=True, alpha=0.6, color='c')

# Add a line for the probability density function (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pdf = lambda_ * np.exp(-lambda_ * x)
plt.plot(x, pdf, 'r-', lw=2)

# Labeling the plot
plt.title(f"Exponential Distribution (lambda={lambda_})")
plt.xlabel('Value')
plt.ylabel('Density')
plt.xlim(0, 20)
plt.grid(True)

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Parameters for the 2D normal distribution
mu_X = 0  # Mean of X
mu_Y = 0  # Mean of Y
sigma_X = 1  # Standard deviation of X
sigma_Y = 1  # Standard deviation of Y
rho = 0.8  # Correlation between X and Y

# Covariance matrix for the joint distribution
cov_matrix = [[sigma_X**2, rho*sigma_X*sigma_Y],
              [rho*sigma_X*sigma_Y, sigma_Y**2]]

# Create a meshgrid for the values of X and Y
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

# Simulate the joint distribution f(x, y)
rv = multivariate_normal([mu_X, mu_Y], cov_matrix)
f_xy = rv.pdf(pos)

# Plot the joint density f(x, y)
plt.figure(figsize=(14, 6))

# First plot: Joint density f(x, y)
plt.subplot(1, 2, 1)
plt.contourf(X, Y, f_xy, cmap='viridis')
plt.colorbar(label='Density')
plt.title('Joint Density f(x, y)')
plt.xlabel('X')
plt.ylabel('Y')

# Conditional density f(y|x) for a fixed value of X (e.g., X = 1)
x_fixed = 1
mean_cond = mu_Y + rho * (sigma_Y / sigma_X) * (x_fixed - mu_X)
var_cond = sigma_Y**2 * (1 - rho**2)
f_y_given_x = (1 / np.sqrt(2 * np.pi * var_cond)) * np.exp(- (y - mean_cond)**2 / (2 * var_cond))

# Second plot: Conditional density f(y|x) for fixed X = 1
plt.subplot(1, 2, 2)
plt.plot(y, f_y_given_x, label=f'Conditional density f(y|x) for X = {x_fixed}', color='orange')
plt.title(f'Conditional Density f(y|x) for X = {x_fixed}')
plt.xlabel('Y')
plt.ylabel('Density')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Ÿæÿßÿ±ÿßŸÖÿ™ÿ±Ÿáÿß
p = 0.05  # ÿßÿ≠ÿ™ŸÖÿßŸÑ ŸÖÿπ€åŸàÿ® ÿ®ŸàÿØŸÜ €å⁄© ŸÇÿ∑ÿπŸá (5%)
n = 100   # ÿ™ÿπÿØÿßÿØ ŸÇÿ∑ÿπÿßÿ™ ÿØÿ± Ÿáÿ± ÿØÿ≥ÿ™Ÿá ÿ™ŸàŸÑ€åÿØ
k = np.arange(0, n+1)  # ÿ™ÿπÿØÿßÿØ ŸÇÿ∑ÿπÿßÿ™ ŸÖÿπ€åŸàÿ® ÿØÿ± ÿØÿ≥ÿ™Ÿá‚ÄåŸáÿß (ÿßÿ≤ 0 ÿ™ÿß n)

# ŸÖÿ≠ÿßÿ≥ÿ®Ÿá ÿ™Ÿàÿ≤€åÿπ ÿ®ÿß€åŸÜŸàŸÖ€åÿßŸÑ ÿ®ÿ±ÿß€å ÿ™ÿπÿØÿßÿØ ŸÖÿπ€åŸàÿ®‚ÄåŸáÿß
pmf = binom.pmf(k, n, p)

# ŸÜŸÖÿß€åÿ¥ ⁄Øÿ±ÿßŸÅ ÿ™Ÿàÿ≤€åÿπ ÿßÿ≠ÿ™ŸÖÿßŸÑ ÿ®ÿß€åŸÜŸàŸÖ€åÿßŸÑ
plt.figure(figsize=(10,6))
plt.vlines(k, 0, pmf, colors='b', lw=5)
plt.plot(k, pmf, 'bo', ms=8)
plt.title(f'Tozi\'e Binaomi\'al baraye {n} ghesmat', fontsize=16)
plt.xlabel('Shomare ghesmat mo\'ayeb', fontsize=12)
plt.ylabel('Ihtemal', fontsize=12)
plt.grid(True)
plt.show()

# ŸÖÿ≠ÿßÿ≥ÿ®Ÿá ÿßÿ≠ÿ™ŸÖÿßŸÑ ÿ±ÿØ ⁄©ÿ±ÿØŸÜ ÿØÿ≥ÿ™Ÿá‚Äåÿß€å ÿßÿ≤ ŸÇÿ∑ÿπÿßÿ™ ÿß⁄Øÿ± ÿ™ÿπÿØÿßÿØ ŸÇÿ∑ÿπÿßÿ™ ŸÖÿπ€åŸàÿ® ÿ®€åÿ¥ ÿßÿ≤ ÿ≠ÿØ ŸÖÿπ€åŸÜ ÿ®ÿßÿ¥ÿØ
threshold = 10  # ÿ™ÿπÿØÿßÿØ ŸÖÿπ€åŸàÿ® ŸÖÿ¨ÿßÿ≤
probability_of_rejecting = 1 - binom.cdf(threshold, n, p)
print(f"Emkan reftan daste agar {threshold} ghesmat mo'ayeb bish az in bashad: {probability_of_rejecting:.4f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# ÿ™ŸàŸÑ€åÿØ ÿØÿßÿØŸá‚ÄåŸáÿß€å ŸÖÿµŸÜŸàÿπ€å ÿ®ÿ±ÿß€å ÿß€åŸÜ ŸÖÿ´ÿßŸÑ
np.random.seed(42)

# ÿ™ÿπÿØÿßÿØ ŸÇÿ∑ÿπÿßÿ™
n_samples = 1000

# Ÿà€å⁄ò⁄Ø€å‚ÄåŸáÿß€å ÿ™ŸàŸÑ€åÿØ (ŸÖÿ´ŸÑÿßŸã ÿØŸÖÿßÿå ŸÅÿ¥ÿßÿ±ÿå ÿ≥ÿ±ÿπÿ™ ÿ™ŸàŸÑ€åÿØ)
temperature = np.random.normal(loc=25, scale=5, size=n_samples)  # ÿØŸÖÿß (ÿØÿ±ÿ¨Ÿá ÿ≥ÿßŸÜÿ™€å‚Äå⁄Øÿ±ÿßÿØ)
pressure = np.random.normal(loc=30, scale=3, size=n_samples)  # ŸÅÿ¥ÿßÿ± (ÿ®ÿßÿ±)
speed = np.random.normal(loc=100, scale=10, size=n_samples)  # ÿ≥ÿ±ÿπÿ™ ÿ™ŸàŸÑ€åÿØ (ÿØŸàÿ± ÿØÿ± ÿØŸÇ€åŸÇŸá)

# ŸÖÿ™ÿ∫€åÿ± ŸáÿØŸÅ (0 ÿ≥ÿßŸÑŸÖÿå 1 ŸÖÿπ€åŸàÿ®)
# ŸÅÿ±ÿ∂ ŸÖ€å‚Äå⁄©ŸÜ€åŸÖ ⁄©Ÿá ÿØŸÖÿßÿå ŸÅÿ¥ÿßÿ± Ÿà ÿ≥ÿ±ÿπÿ™ ÿ™ŸàŸÑ€åÿØ ÿ™ÿ£ÿ´€åÿ± ŸÖÿ≥ÿ™ŸÇ€åŸÖ ÿ®ÿ± ŸÖÿπ€åŸàÿ® ÿ®ŸàÿØŸÜ ŸÇÿ∑ÿπŸá ÿØÿßÿ±ŸÜÿØ
def is_defective(temp, pres, speed):
    return (temp > 28) & (pres < 25) | (speed > 110)

defective = is_defective(temperature, pressure, speed)

# ÿ≥ÿßÿÆÿ™ DataFrame
data = pd.DataFrame({
    'temperature': temperature,
    'pressure': pressure,
    'speed': speed,
    'defective': defective
})

# ÿ™ŸÇÿ≥€åŸÖ ÿØÿßÿØŸá‚ÄåŸáÿß ÿ®Ÿá Ÿà€å⁄ò⁄Ø€å‚ÄåŸáÿß Ÿà ŸÖÿ™ÿ∫€åÿ± ŸáÿØŸÅ
X = data[['temperature', 'pressure', 'speed']]  # Ÿà€å⁄ò⁄Ø€å‚ÄåŸáÿß
y = data['defective']  # ŸáÿØŸÅ

# ÿ™ŸÇÿ≥€åŸÖ ÿØÿßÿØŸá‚ÄåŸáÿß ÿ®Ÿá ÿØÿßÿØŸá‚ÄåŸáÿß€å ÿ¢ŸÖŸàÿ≤ÿ¥€å Ÿà ÿ¢ÿ≤ŸÖÿß€åÿ¥€å
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ÿßÿ≥ÿ™ÿßŸÜÿØÿßÿ±ÿØÿ≥ÿßÿ≤€å ÿØÿßÿØŸá‚ÄåŸáÿß
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# ŸÖÿØŸÑ ÿØÿ±ÿÆÿ™ ÿ™ÿµŸÖ€åŸÖ
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ÿ±Ÿà€å ÿØÿßÿØŸá‚ÄåŸáÿß€å ÿ¢ÿ≤ŸÖÿß€åÿ¥€å
y_pred = model.predict(X_test)

# ÿßÿ±ÿ≤€åÿßÿ®€å ŸÖÿØŸÑ
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# ŸÜŸÖÿß€åÿ¥ ŸÜÿ™ÿß€åÿ¨
print(f"Accuracy: {accuracy:.4f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)

# ŸÜŸÖÿß€åÿ¥ ⁄Øÿ±ÿßŸÅ ÿØÿ±ÿÆÿ™ ÿ™ÿµŸÖ€åŸÖ
from sklearn.tree import plot_tree
plt.figure(figsize=(12,8))
plot_tree(model, filled=True, feature_names=['temperature', 'pressure', 'speed'], class_names=['Not Defective', 'Defective'], rounded=True)
plt.show()
```

```python
# Ÿàÿßÿ±ÿØ ⁄©ÿ±ÿØŸÜ ⁄©ÿ™ÿßÿ®ÿÆÿßŸÜŸá‚ÄåŸáÿß
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# ÿ®ÿßÿ±⁄Øÿ∞ÿßÿ±€å ÿØ€åÿ™ÿßÿ≥ÿ™ Iris
iris = load_iris()
X = iris.data  # Ÿà€å⁄ò⁄Ø€å‚ÄåŸáÿß (ÿ∑ŸàŸÑ Ÿà ÿπÿ±ÿ∂ ⁄ØŸÑÿ®ÿ±⁄Ø Ÿà ÿ≥ÿßŸÇŸá)
y = iris.target  # ÿ®ÿ±⁄Üÿ≥ÿ®‚ÄåŸáÿß (ŸÜŸàÿπ ⁄ØŸÑ)

# ÿ™ŸÇÿ≥€åŸÖ ÿØÿßÿØŸá‚ÄåŸáÿß ÿ®Ÿá ÿØÿßÿØŸá‚ÄåŸáÿß€å ÿ¢ŸÖŸàÿ≤ÿ¥€å Ÿà ÿ¢ÿ≤ŸÖÿß€åÿ¥€å
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ÿßÿ≥ÿ™ÿßŸÜÿØÿßÿ±ÿØÿ≥ÿßÿ≤€å ÿØÿßÿØŸá‚ÄåŸáÿß
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# ŸÖÿØŸÑ ÿ±⁄Øÿ±ÿ≥€åŸàŸÜ ŸÑÿ¨ÿ≥ÿ™€å⁄©
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)

# Ÿæ€åÿ¥‚Äåÿ®€åŸÜ€å ÿ®ÿ± ÿ±Ÿà€å ÿØÿßÿØŸá‚ÄåŸáÿß€å ÿ¢ÿ≤ŸÖÿß€åÿ¥€å
y_pred = model.predict(X_test)

# ÿßÿ±ÿ≤€åÿßÿ®€å ŸÖÿØŸÑ
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# ŸÜŸÖÿß€åÿ¥ ŸÜÿ™ÿß€åÿ¨
print(f"Accuracy: {accuracy:.4f}")
print("\nConfusion Matrix:")
print(conf_matrix)
print("\nClassification Report:")
print(class_report)

# ŸÜŸÖÿß€åÿ¥ ⁄Øÿ±ÿßŸÅ ŸÖÿßÿ™ÿ±€åÿ≥ ÿßÿ¥ÿ™ÿ®ÿßŸá
plt.figure(figsize=(6,6))
plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
classes = iris.target_names
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes, rotation=45)
plt.yticks(tick_marks, classes)
plt.tight_layout()
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
