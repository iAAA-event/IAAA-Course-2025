### üöÄüì¶ Installing and Importing Packages

Before we start working with data, we need two important Python libraries:

### 1Ô∏è‚É£ **Pandas** üêº  
- **What it is:** Pandas is a powerful library for **data manipulation and analysis**.  

- **Why we use it:**

### üì• Creating and Loading a Dataset



```python
data = {
    "Product": ["Laptop", "Phone", "Tablet", "Headphones", "Monitor", "Keyboard", "Mouse", "Printer", "Camera", "Smartwatch"],
    "Category": ["Electronics", "Electronics", "Electronics", "Accessories", "Electronics", "Accessories", "Accessories", "Electronics", "Electronics", "Electronics"],
    "Sales": [1200, 800, 600, 300, 450, 150, 100, 350, 500, 700],
    "Quantity": [5, 10, 7, 15, 6, 20, 25, 4, 8, 9],
    "Rating": [4.5, 4.2, 4.0, 3.8, 4.1, 3.9, 3.7, 4.3, 4.4, 4.0]
}

df = pd.DataFrame(data)

df.to_csv("products.csv", index = False)

df = pd.read_csv("products.csv")
```

### üí° Explanation

- to_csv() saves the table into a CSV file so we can simulate loading data from an external source.

- pd.read_csv() loads the CSV file back into Pandas.

## üìå Section 2: Visualization with Matplotlib Library

### üé® Basic Plots in Python

Data visualization helps us understand data quickly and identify patterns, trends, or anomalies.

Here, we‚Äôll cover the most commonly used plots in Python for beginners.

1. Line Plot üìà

- Purpose: Shows the change of a variable over time or an index.

- Applications: Trends, time series, continuous data.

- Advantages: Makes patterns and trends easy to see.

```python
plt.figure(figsize=(8,5))
plt.plot(df['Product'], df['Sales'], marker='o', color='blue')
plt.xlabel('Product')
plt.ylabel('Sales ($)')
plt.title('Line Plot: Sales per Product')
plt.xticks(rotation=45)
plt.show()
```

### üí° Explanation

- Each point represents the sales of a product.

- marker='o' marks each data point.

- xticks(rotation=45) rotates the product labels for readability.

2. Vertical Bar Chart üìä

- Purpose: Compare quantities between different categories.

- Applications: Grouped data, categorical comparisons.

- Advantages: Shows clear differences between groups.

```python
plt.figure(figsize=(8,5))
plt.bar(df['Product'], df['Sales'], color='orange', edgecolor='black')
plt.xlabel('Product')
plt.ylabel('Sales ($)')
plt.title('Bar Chart: Sales per Product')
plt.xticks(rotation=45)
plt.show()
```

### üí° Explanation

- Vertical bars represent sales of each product.

- edgecolor='black' adds borders for better visibility.

3. Horizontal Bar Chart üìä

```python
plt.figure(figsize=(8,5))
plt.barh(df['Product'], df['Quantity'], color='lightgreen', edgecolor='black')
plt.xlabel('Quantity Sold')
plt.ylabel('Product')
plt.title('Horizontal Bar Chart: Quantity Sold per Product')
plt.show()
```

### üí° Explanation

- Horizontal bars are useful when category labels are long.

- Easy to compare quantities across products.

4. Histogram üìâ

- Purpose: Show the distribution of a numeric variable.

- Applications: Understand how data is spread, detect skewness or outliers.

```python
plt.figure(figsize=(8,5))
plt.hist(df['Sales'], bins=5, color='purple', edgecolor='black')
plt.xlabel('Sales ($)')
plt.ylabel('Frequency')
plt.title('Histogram: Sales Distribution')
plt.show()
```

### üí° Explanation

- bins=5 divides the range of values into 5 intervals.

- Useful for observing which sales ranges are most common.

5. Scatter Plot ‚ö™

- Purpose: Visualize the relationship between two numeric variables.

- Applications: Correlation, pattern detection, outliers.

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Quantity'], df['Sales'], color='red', edgecolors='black')
plt.xlabel('Quantity Sold')
plt.ylabel('Sales ($)')
plt.title('Scatter Plot: Sales vs Quantity')
plt.show()
```

### üí° Explanation

- Each point represents a product.

- X-axis = Quantity, Y-axis = Sales.

- Helps identify trends or unusual data points.



6. Pie Chart üçï

- Purpose: Show relative proportions of categories.

- Applications: Market share, percentage breakdowns.

- Advantages: Visualizes contribution of each group to the whole.

```python
category_sales = df.groupby('Category')['Sales'].sum()

plt.figure(figsize=(6,6))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=90, colors=['lightblue','lightcoral'])
plt.title('Pie Chart: Sales Share by Category')
plt.show()
```

### üí° Explanation

- Each slice represents a category‚Äôs contribution to total sales.

- autopct='%1.1f%%' shows percentage values.
